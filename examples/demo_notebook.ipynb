The widgets extend the DOMWidget class from ipywidgets and should thus behave like regular jupyter widgets.

To show a widget, import it, instance it and call display(widget_ref).

```python
# Import necessary libraries
import numpy as np
from ipy_opengl_utils.particle_widget import ParticleWidget

# Set up the OpenGL widget
width, height = 800, 600
widget = ParticleWidget(width, height, draw_axes=True, select_particles=False)

count = 100
test_positions = (np.random.rand(count, 3).astype(np.float32) - 0.5) * 10
test_colors = np.ones_like(test_positions)
widget.add_particles(test_positions, test_colors, update_cam=True)

display(widget)
```
